```python
import pandas as pd
import numpy as np

df = pd.read_csv("data\salesdaily.csv")
print(df.columns)
```

1. Total sales quantities for each drug category (ATC code)

```python
drugs = ['M01AB', 'M01AE', 'N02BA', 'N02BE', 'N05B', 'N05C', 'R03','R06']
sales = df[drugs].sum().sort_values(ascending = False)
print(sales)
```

2. Which individual drug brands have the highest total sales?

```python
highest_drug = df[drugs].sum().idxmax()

print(highest_drug)
```

3. Which three drugs have the highest sales in January 2015, July 2016, September 2017.

```python
jan = df[(df["Year"] == 2015) & (df["Month"] == 1)]
janSales = jan[drugs].sum().sort_values(ascending = False).head(3)
print("Highest in Jan 2015: ")
print(janSales)

july = df[(df["Year"] == 2016) & (df["Month"] == 7)]
julSales = july[drugs].sum().sort_values(ascending = False).head(3)
print("Highest in July 2016: ")
print(julSales)

sep = df[(df["Year"] == 2017) & (df["Month"] == 9)]
sepSales = sep[drugs].sum().sort_values(ascending = False).head(3)
print("Highest in September 2017: ")
print(sepSales)
```

Which drug has sold the most often in 2017?

```python
sales = df[(df["Year"] == 2017)]
s = sales[drugs].sum().sort_values(ascending = False).head(1)
print(s.idxmax())
```

5. Which drug category has the highest average daily sales?

```python
avg = df[drugs].mean().sort_values(ascending = False).head(1)
print(avg)
```

6. Are respiratory drugs (R03) sold more during specific months?

```python
bymonths = df["R03"].groupby(df['Month']).sum().sort_values(ascending = False)
print(bymonths)
```
